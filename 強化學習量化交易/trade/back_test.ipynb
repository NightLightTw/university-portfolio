```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DataFrame = pd.read_csv('./data/tsml.csv')
fig = plt.figure(figsize=(5, 5))
AX  = fig.add_axes([0.1,0.5,1,0.2])
AX2 = fig.add_axes([0.1,0.1,1,0.2])

#表格標題
AX.set_title("close")
AX2.set_title('volumn')

#收盤價
AX.plot(DataFrame['Close'], label='Close', color='k')
#成交量
AX2.bar(DataFrame.index, DataFrame['Volume'], label='Volume', color='c')

AX.legend();
AX2.legend();
```

```python
df_ppo_mlp = pd.read_csv('profit_PPO_MLP.txt',sep=' ',header=None)
```

```python
n = len(df_ppo_mlp[0])
total = sum(df_ppo_mlp[0])
avg_profit = total/n
print('n:',n)
print('total:',total)
print('avg_profit:',avg_profit)
```

```python
turns = []
for i in range(n):
    turns.append(i)
plt.plot(turns,df_ppo_mlp[0])
plt.show()
```
