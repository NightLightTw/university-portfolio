```python
import pandas_datareader as data_reader
import datetime as datetime
import pandas as pd

def dataset_loader(stock_number):
  
    start = datetime.datetime(2012,1,1)
    end   = datetime.datetime(2022,5,28)
    dataset = data_reader.DataReader(stock_number, 'yahoo', start, end)
  
  
    #close = dataset['Close']
    return dataset

stock_number = "2330.TW"
dataset = dataset_loader(stock_number)
data = dataset.drop(['Adj Close'],axis=1)
data
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(5, 5))
AX  = fig.add_axes([0.1,0.5,1,0.2])
AX2 = fig.add_axes([0.1,0.1,1,0.2])
DataFrame=data
#表格標題
AX.set_title("close")
AX2.set_title('volumn')

#收盤價
AX.plot(DataFrame['Close'], label='Close', color='k')
#成交量
AX2.bar(DataFrame.index, DataFrame['Volume'], label='Volume', color='c')

AX.legend();
AX2.legend();
```

```python
df=pd.DataFrame(data)
df.to_csv('tsml.csv',index=1)
```
